```python
from kafka import KafkaConsumer
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Create Kafka consumer
consumer = KafkaConsumer(
    'nextbike_data',  # Use the correct topic name
    bootstrap_servers='localhost:29092',
    value_deserializer=lambda v: json.loads(v.decode('utf-8')),
    auto_offset_reset='earliest'
)
```

```python
target_countries = ["Germany", "Austria"]
max_cities_per_country = 20

def consume_data(max_messages=100):
    print("Starting to consume messages...")

    country_city_bike_counts = {}

    message_count = 0 

    for message in consumer:
        data = message.value

        # Check if 'cities' key exists in the json
        if 'cities' in data:
            country_name = data.get('country_name', 'Unknown Country')

            # Process only the target countries, Germany, Austria & Spain in our case
            if country_name in target_countries:
                # Aggregate data by country and city
                if country_name not in country_city_bike_counts:
                    country_city_bike_counts[country_name] = {}

                for city in data['cities']:
                    city_name = city.get('name', 'Unknown City')
                    available_bikes = sum([place.get('bikes', 0) for place in city.get('places', [])])

                    if city_name in country_city_bike_counts[country_name]:
                        country_city_bike_counts[country_name][city_name] += available_bikes
                    else:
                        country_city_bike_counts[country_name][city_name] = available_bikes
        else:
            print("No 'cities' key found")

        message_count += 1
        if message_count >= max_messages:
            break

    print("Finished consuming messages.")

    # Plot data for each of the specified countries
    for country, cities in country_city_bike_counts.items():
        # Sort cities by available bikes and limit the number of cities
        sorted_cities = sorted(cities.items(), key=lambda item: item[1], reverse=True)[:max_cities_per_country]
        df = pd.DataFrame(sorted_cities, columns=['city', 'available_bikes'])

        print(f"{country}:", df)

        plt.figure(figsize=(15, 7))
        bars = plt.bar(df['city'], df['available_bikes'])

        plt.xticks(rotation=45, ha='right')
        
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, int(yval), ha='center', va='bottom')

        plt.xlabel('Cities', fontsize=14)
        plt.ylabel('Number of Available Bikes', fontsize=14)
        plt.title(f'Available Bikes by City in {country}', fontsize=16)
        plt.tight_layout()
        plt.show()

consume_data(max_messages=200)
```
