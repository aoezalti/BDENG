```python
import requests

# REST API URL
api_url = 'https://api.nextbike.net/maps/nextbike-live.json'

def fetch_data_for_vienna():
    try:
        response = requests.get(api_url)
        response.raise_for_status()  # Fehler bei ungültigem Statuscode auslösen

        data = response.json()

        # Extrahiere aktive und verfügbare Fahrräder mit ihren Koordinaten für Wien
        for country in data.get('countries', []):
            for city in country.get('cities', []):
                if city.get('name') == 'Wien':
                    print(f"Daten für Stadt Wien:")
                    for place in city.get('places', []):
                        for bike_info in place.get('bike_list', []):
                            if bike_info.get('active', False):
                                print(f"Station: {place.get('name')}")
                                print(f"Koordinaten: {place.get('lat')}, {place.get('lng')}")
                                print()

        print(f"Datenabruf für Wien abgeschlossen.")

    except requests.exceptions.RequestException as e:
        print(f"Fehler beim Abrufen der Daten: {e}")

    except Exception as ex:
        print(f"Unerwarteter Fehler: {ex}")

# Funktion aufrufen, um Daten für Wien abzurufen
fetch_data_for_vienna()
```
